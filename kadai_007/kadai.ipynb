```python
import pandas as pd

# csvファイルをデータフレームとして読み込む
df = pd.read_csv("sample_pandas_6.csv")
# 先頭から5行目までを表示する
df.head()
```

```python
category_df = pd.read_csv('category.csv')
category_df
```

```python
df = pd.merge(df, category_df[['商品番号', 'カテゴリー']], how='inner', on='商品番号')
df.head()
```

```python
hoge = df['カテゴリー'].value_counts()
hoge
```

```python
hoge.index
```

```python
plt.bar(hoge.index.tolist(), hoge.values.tolist())
plt.show()
```
